```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

age = np.array([3, 4, 5, 6, 7, 8, 9, 11, 12, 14, 15, 16, 17])
wing_length = np.array([1.4, 1.5, 2.2, 2.4, 3.1, 3.2, 3.2, 3.9, 4.1, 4.7, 4.5, 5.2, 5.0])

#plt.scatter(age, wing_length)
plt.plot(age, wing_length, 'o')
plt.xlabel('age')
plt.ylabel('wing length')

m, b = np.polyfit(age, wing_length, 1)
plt.plot(age, m*age+b)

corr_coef = np.corrcoef(age, wing_length)
print(f'Pearson correlation coeffcient: {corr_coef}')

age = np.array([3, 4, 5, 6, 7, 8, 9, 11, 12, 14, 15, 16, 17]).reshape((-1, 1))

model = LinearRegression().fit(age, wing_length)

r_sq = model.score(age, wing_length)
print(f"Coefficient of determination: {r_sq}")
print(f"Intercept: {model.intercept_}")
print(f"Slope: {model.coef_}")

```
